```python
%matplotlib widget
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries = pd.read_csv('Countries Data.csv')
countries.head()
```

```python
c_52 = countries.loc[countries['year'] == 1952]
c_52.head()
```

```python
c_2007 = countries.loc[countries['year'] == 2007]
c_2007.head()
```

```python
c_merge = c_52.merge(c_2007, left_on='country', right_on='country')
c_merge.head()
```

```python
drop_df = c_merge.drop(['year_x', 'year_y'], axis=1)
drop_df.head()
```

```python
c_merge['population_growth'] = c_merge['population_y'] - c_merge['population_x']
c_merge.head()
```

```python
c_merge = c_merge.sort_values('population_growth', ascending=False).head(10)
c_merge.head(10)
```

```python
names = ['China', 'India', 'United States', 'Indonesia', 'Brazil', 'Pakistan', 'Bangladesh', 'Nigeria', 'Mexico', 'Philippines']
pop_grow = (c_merge['population_growth'] / 10**6)
```

```python
plt.figure(figsize=(11,9))
plt.bar(names,pop_grow,width=0.6)
plt.title('Population Growth')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```
