```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0,5,10,15,20,25,30]
y1 = [10,15,20,20,30,15,0]
y2 = [10,12,15,12,20,10,0]
```

```python
plt.title('Vilosity-Time Graph')
plt.xlabel('Vilocity(m/s)')
plt.ylabel('Time(s)')
plt.plot(x,y1,linestyle='dashed',marker='d',label='Plot-1')
plt.plot(x,y2,linestyle='dotted',marker='*',label='Plot-2')
plt.xlim(5,25)
plt.ylim(5,25)
plt.legend()
plt.show()
```

```python
x=np.arange(1,11,1)
print(x)

a1=(2*x)+1
a2=(2*x**2)+3
```

```python
plt.plot(x,a1,color='g',linewidth=3,label='y = 2x+1')
plt.plot(x,a2,color='r',linewidth=3,label='y = 2x^2+3')
plt.legend()
plt.grid(True)
plt.show()
```
