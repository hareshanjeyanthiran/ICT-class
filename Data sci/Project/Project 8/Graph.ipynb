```python
%matplotlib widget
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
births_july = [12,15,11,19,21,17,2]

df_july = pd.DataFrame({
    "Day": days,
    "New Births": births_july
})

print(df_july)
```

```python
plt.figure()
plt.plot(df_july["Day"],
         df_july["New Births"],
         marker='o',
         linestyle='--',
         linewidth=2,
         color='blue')
plt.title("Birth Rate - July")
plt.xlabel("Day")
plt.ylabel("New Births")
plt.show()
```

```python
births_aug = [11,8,29,14,18,10,9]
df_aug = pd.DataFrame({
    "Day": days,
    "New Births": births_aug
})
print(df_aug)
```

```python
plt.figure()
plt.plot(df_aug["Day"],
         df_aug["New Births"],
         marker='s',
         linestyle='-.',
         linewidth=2,
         color='green')
plt.title("Birth Rate - August")
plt.xlabel("Day")
plt.ylabel("New Births")
plt.show()
```

```python
plt.figure()
plt.plot(days, births_july,
         marker='o',
         linestyle='--',
         linewidth=2,
         label="July")
plt.plot(days, births_aug,
         marker='s',
         linestyle='-.',
         linewidth=2,
         label="August")
plt.title("Birth Rate Comparison")
plt.xlabel("Day")
plt.ylabel("New Births")
plt.legend()
plt.show()
```

```python
x = np.arange(1,11)
y = 6*x**2 + x + 1

plt.figure()
plt.plot(x, y,
         marker='o',
         linestyle='-',
         linewidth=2)
plt.title("Graph of y = 6x² + x + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
