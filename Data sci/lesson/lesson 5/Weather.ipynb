```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv('Weather.csv')
weather.head()
```

```python
weather.info()
```

```python
sns.barplot(x=weather['weather_type'], y=weather['temperature'], color='blue')
```

```python
sns.histplot(weather['humidity'], color='green', kde=True)
```

```python
sns.histplot(weather['humidity'], kde=False)  
sns.rugplot(weather['humidity'], color='darkblue')  
plt.show()
```

```python
sns.jointplot(x=weather['humidity'], y=weather['temperature'], color='cyan')
```

```python
sns.jointplot(x=weather['humidity'], y=weather['temperature'], color='lightblue',kind="hex")
```

```python
sns.jointplot(x=weather['humidity'], y=weather['temperature'], color='darkblue',kind="kde")
```

```python
sns.pairplot(weather[['humidity','temperature','air_pollution_index']])
```

```python
sns.stripplot(x=weather['weather_type'], y=weather['temperature'], palette='summer')
```

```python
sns.swarmplot(x=weather['weather_type'], y=weather['temperature'], palette='winter')
```

```python
sns.countplot(x=weather['weather_type'])
```

```python
sns.pointplot(x=weather['humidity'], y=weather['temperature'], hue=weather['weather_type'])
```

```python
sns.boxplot(x=weather['weather_type'], y=weather['temperature'])
```

```python
sns.lmplot(x='humidity', y='temperature', data=weather)
```
